```python
df.drop(['Date','Location'],axis=1,inplace=True)
```

```python
cat_fut=[col for col in df.columns if df[col].dtype=='O']
num_fut=[col for col in df.columns if df[col].dtype!='O']
print('categorical futures:',cat_fut)
print('numerical futures:',num_fut)
```

```python
#handling missing values
for i in num_fut:
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
for i in cat_fut:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
#outliers detection
def detect_outliers(data):
    out=[]
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    IQR=q3-q1
    lwr_bound=q1-(IQR*1.5)
    upper_bound=q3+(IQR*1.5)
    
    for i in data:
        if i<lwr_bound or i>upper_bound:
            out.append(i)
    return out

for i in num_fut:
    print('outliers of ',i,' attribute:',len(detect_outliers(df[i])))
    
```

```python
#outliers handking
def handle_outliers(data):
    out=[]
    q1=data.quantile(0.25)
    q3=data.quantile(0.75)
    IQR=q3-q1
    lwr_bound=q1-(IQR*1.5)
    upper_bound=q3+(IQR*1.5)
    
    b=np.where(data<lwr_bound,lwr_bound,data)
    b1=np.where(b>upper_bound,upper_bound,b)
    return b1
for i in num_fut:
    df[i]=handle_outliers(df[i])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import NuSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve,classification_report,plot_roc_curve,roc_auc_score,precision_score,recall_score,f1_score
```

```python
matrix=np.triu(df.corr())
plt.figure(figsize=(20,9))
sns.heatmap(df.corr(),annot=True,cmap='bwr_r',mask=matrix)
```

```python
y=df['RainTomorrow']
X=df.drop(['RainTomorrow'],axis=1)
```

```python
from keras import utils
```

```python
#under sampling
# from imblearn.under_sampling import RandomUnderSampler
# rus = RandomUnderSampler()
# X_res,y_res=rus.fit_resample(X,y)
# print(df['RainTomorrow'].value_counts())
# print(X_res.shape,y_res.shape)
```

```python
#oversampling
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_res,y_res= ros.fit_resample(X,y)
print(df['RainTomorrow'].value_counts())
print(X_res.shape,y_res.shape)
```

```python
# Implementing Oversampling for Handling Imbalanced 
# from imblearn.combine import SMOTETomek
# smk = SMOTETomek(random_state=42)
# X_res,y_res=smk.fit_resample(X,y)
# print(df['RainTomorrow'].value_counts())
# print(X_res.shape,y_res.shape)

```

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.2,random_state=2)
```

# Decision tree classifier

```python
CA=DecisionTreeClassifier()
CA=CA.fit(X_train,y_train)
y_pred=CA.predict(X_test)

c_matrix=confusion_matrix(y_test,y_pred)
acc_1=round(accuracy_score(y_test,y_pred)*100,2)
cls_rep=classification_report(y_test,y_pred)
print('Confusion Matrix:\n',c_matrix,'\nAccuracy:',acc_1,'\nClassification Report:\n',cls_rep)

dt_precision = round(precision_score(y_test, y_pred)*100,2)
dt_recall = round(recall_score(y_test, y_pred)*100,2)
dt_fScore=round(f1_score(y_test, y_pred)*100,2)


print('\nPrecision: ',dt_precision,'%')
print('\nRecall: ',dt_recall,'%')
print('\nF1 Score:',dt_fScore,'%')
```

```python
plot_roc_curve(CA, X_test, y_test)
roc_auc_score(y_test, y_pred, average=None) 
plt.title('ROC curve for Decision Tree Classifier Model')
plt.xlabel("False Positive Rate")
plt.ylabel('True Positive Rate')
```
